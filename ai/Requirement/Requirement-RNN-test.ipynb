```python
import pandas as pd

df = pd.read_csv('datasets/images/results.csv',sep='|')
df.head()
```

```python
# 캡션데이터 받아오기
text = df[' comment']
text = text.dropna()
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
# corpus 단어 토큰화
t = Tokenizer()

#전체 corpus 빈도기반 tokenizing 수행 -> 빈도가 높을수록 높은 번호를 부여
#fit_on_texts는 pandas와도 호환 가능
t.fit_on_texts(text)

vocab_size = len(t.word_index) + 1
vocab_size
```

```python
t.word_index
```

```python
# 캡션을 vector로 변환
encoded = t.texts_to_sequences(text)
```

```python
# 캡션 최대길이 반환
max_len=max(len(l) for l in encoded) # 모든 샘플에서 길이가 가장 긴 샘플의 길이 출력
print('샘플의 최대 길이 : {}'.format(max_len))
```

```python
# 샘플 최대길이에 맞게 padding
encoded = pad_sequences(encoded, maxlen=max_len, padding='pre')
encoded
```
